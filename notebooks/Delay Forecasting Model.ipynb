```python
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
```

## Loading inputs and outputs for forecasting model

```python
x = pd.read_csv('../data/model_inputs.csv')
y = pd.read_csv('../data/model_outputs.csv')

print('INPUTS:')
x.info()
print('\nOUTPUTS:')
y.info()
```

## Format data for model

**Use data from 2003 to 2018 for training and test on the data from 2019 and 2020**

```python
enc = OneHotEncoder(handle_unknown='ignore')

carrier_enc = enc.fit_transform(x[['carrier_name']].values).toarray()
airport_enc = enc.fit_transform(x[['airport_name']].values).toarray()
month_enc = enc.fit_transform(x[['month']].values).toarray()
year = x[['year']].values-2003

nas_delay_prob = y[['nas_delay_prob']].values
carrier_delay_prob = y[['carrier_delay_prob']].values
nas_delay = y[['nas_delay']].values
carrier_delay = y[['carrier_delay']].values

xdata = np.hstack((year,month_enc,airport_enc,carrier_enc))
ydata = np.hstack((nas_delay_prob,nas_delay,carrier_delay_prob,carrier_delay))

train_idxs = (x['year']<2019).values
test_idxs_2019 = (x['year']==2019).values
test_idxs_2020 = (x['year']==2020).values

xtrain = xdata[train_idxs]
ytrain = ydata[train_idxs]

xtest_2019 = xdata[test_idxs_2019]
ytest_2019 = ydata[test_idxs_2019]

xtest_2020 = xdata[test_idxs_2020]
ytest_2020 = ydata[test_idxs_2020]
```

## Train linear regression model to predict delay rate and time

```python
# ONLY run this cell if you want to save the trained model
rf_model = RandomForestRegressor(random_state=0)
rf_model.fit(xtrain, ytrain)
with open('../models/rf_model.pkl', 'wb') as fp:
    pickle.dump(rf_model,fp)
```

## Measure linear fit of the model on train and test data

```python
# load RF model trained with the training data
with open('../models/rf_model.pkl', 'rb') as fp:
    rf_model = pickle.load(fp)

score_train = rf_model.score(xtrain,ytrain)
print('2003-2018 (Train data) - Coefficient of determination (R2 Score):',score_train)

score_test_2019 = rf_model.score(xtest_2019,ytest_2019)
print('2019 (Test data) - Coefficient of determination (R2 Score):',score_test_2019)

score_test_2020 = rf_model.score(xtest_2020,ytest_2020)
print('2020 (Test data) - Coefficient of determination (R2 Score):',score_test_2020)
```

```python
# Only run this cell if you want to save the predictions of the trained model
for data in [(xtrain,ytrain,train_idxs,'train_results'),
             (xtest_2019,ytest_2019,test_idxs_2019,'test_2019_results'),
             (xtest_2020,ytest_2020,test_idxs_2020,'test_2020_results')]:
    
    y_pred = rf_model.predict(data[0])
    y_error = y_pred[:,i]-data[1][:,i]
    inputs = x.iloc[data[2]].values
    
    results = np.hstack((inputs,data[1],y_pred))
    df = pd.DataFrame(data=results,columns=['nas_delay_prob_true','nas_delay_time_true','carrier_delay_prob_true','carrier_delay_time_true','nas_delay_prob_pred','nas_delay_time_pred','carrier_delay_prob_pred','carrier_delay_time_pred'])
    df.to_csv(data[3]+'.csv', encoding='utf-8', index=False)
```

## Visualize predictions and evaluate forecasting performance

```python
fig, ax = plt.subplots(2,2,figsize=(15,12))

for i in range(2):
    for j in range(2):
        for r,data in enumerate([(xtrain,ytrain,'2003-2018'),
                                 (xtest_2019,ytest_2019,'2019'),
                                 (xtest_2020,ytest_2020,'2020')]):

            y_pred = rf_model.predict(data[0])
            y_error = y_pred[:,i*2+j]-data[1][:,i*2+j]
            if r==0:
                df_temp = pd.DataFrame(data=y_error,columns=[data[2]])
            else:
                d = np.empty(ytrain[:,0].shape)
                d[:] = np.nan
                d[:len(y_error)] = y_error
                df_temp[data[2]] = d

        ax[i,j].boxplot(df_temp.dropna().values,showfliers=False,vert=False)
        ax[i,j].set_yticks(range(1,4))
        ax[i,j].set_yticklabels(df_temp.columns.values)
        ax[i,j].grid(True,alpha=0.6)

        if (i*2+j)%2==0:
            ax[i,j].set_title('NAS Monthly Delay Rate' if i==0 else 'Carrier Monthly Delay Rate')
            ax[i,j].set_xlim(-0.2,0.2)
            ax[i,j].set_xlabel('Prediction Error')
        else:
            ax[i,j].set_title('NAS Monthly Average Delay Time' if i==0 else 'Carrier Monthly Average Delay Time')
            ax[i,j].set_xlim(-100,100)
            ax[i,j].set_xlabel('Prediction Error (minutes)')

plt.show()
```

```python
fig, ax = plt.subplots(3,2,figsize=(15,15))

i = 1 # 0,1,2,3

for r,data in enumerate([(xtrain,ytrain,'2003-2018 (Train data)','royalblue'),
                         (xtest_2019,ytest_2019,'2019 (Test data)','green'),
                         (xtest_2020,ytest_2020,'2020 (Test data)','tomato')]):
    
    y_pred = rf_model.predict(data[0])

    sns.scatterplot(x=data[1][:,i],y=y_pred[:,i],ax=ax[r,0],color=data[3],alpha=0.6,label=data[2])
    ax[r,0].set_ylabel('Predicted delay rate')
    ax[r,0].set_xlabel('Montly delay rate')
    ax[r,0].legend()

    sns.histplot(y_pred[:,i]-data[1][:,i],kde=True,ax=ax[r,1],color=data[3],label=data[2])
    ax[r,1].set_ylabel('Occurrences')
    ax[r,1].set_xlabel('Residual Error')
    ax[r,1].legend()

plt.show()
```

# Proposed approach using selected data

```python
carrier_names = x['carrier_name'].unique()

selected_carriers = []
for i in carrier_names:
    if len(x[x['carrier_name']==i]['year'].unique())==18: 
        selected_carriers.append(i)
        
cond_carriers = x['carrier_name'].isin(selected_carriers)
```

```python
cond_train = x['year']<2019
cond_test2019 = x['year']==2019
cond_test2020 = x['year']==2020

enc = OneHotEncoder(handle_unknown='ignore')

xtrain = []
xtest_2019 = []
xtest_2020 = []

for i in ['year','month','airport_name','carrier_name']:
    
    if i == 'year':
        xtrain.append(x.loc[cond_carriers&cond_train,[i]].values-2003)
        xtest_2019.append(x.loc[cond_carriers&cond_test2019,[i]].values-2003)
        xtest_2020.append(x.loc[cond_carriers&cond_test2020,[i]].values-2003)

    else: 
        xtrain.append(enc.fit_transform(x.loc[cond_carriers&cond_train,[i]].values).toarray())
        xtest_2019.append(enc.fit_transform(x.loc[cond_carriers&cond_test2019,[i]].values).toarray())
        xtest_2020.append(enc.fit_transform(x.loc[cond_carriers&cond_test2020,[i]].values).toarray())
        
xtrain = np.hstack(tuple(xtrain))
xtest_2019 = np.hstack(tuple(xtest_2019))
xtest_2020 = np.hstack(tuple(xtest_2020))

ytrain = []
ytest_2019 = []
ytest_2020 = []

for j in ['nas_delay_prob','nas_delay','carrier_delay_prob','carrier_delay']:
    
    ytrain.append(y.loc[cond_carriers&cond_train,[j]].values)
    ytest_2019.append(y.loc[cond_carriers&cond_test2019,[j]].values)
    ytest_2020.append(y.loc[cond_carriers&cond_test2020,[j]].values)
    
ytrain = np.hstack(tuple(ytrain))
ytest_2019 = np.hstack(tuple(ytest_2019))
ytest_2020 = np.hstack(tuple(ytest_2020))
```

```python
# ONLY run this cell if you want to save the trained model
rf_model = RandomForestRegressor(random_state=0)
rf_model.fit(xtrain, ytrain)
with open('../models/rf_model_for_selected_carriers.pkl', 'wb') as fp:
    pickle.dump(rf_model,fp)
```

```python
# load RF model trained with the training data
with open('../models/rf_model_for_selected_carriers.pkl', 'rb') as fp:
    rf_model = pickle.load(fp)

score_train = rf_model.score(xtrain,ytrain)
print('2003-2018 (Train data) - Coefficient of determination (R2 Score):',score_train)

score_test_2019 = rf_model.score(xtest_2019,ytest_2019)
print('2019 (Test data) - Coefficient of determination (R2 Score):',score_test_2019)

score_test_2020 = rf_model.score(xtest_2020,ytest_2020)
print('2020 (Test data) - Coefficient of determination (R2 Score):',score_test_2020)
```
