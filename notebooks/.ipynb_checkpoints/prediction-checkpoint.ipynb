```python
# Monte-Carlo Simulation : estimates airport runway capacity, not very accurate
# Bayesian and Gaussian mixture model-expectation maximization (GMM-EM) : estimates flight delay in Brazil, good accuracy
# SVM regression : predicts flight delay time in US airports, best results so far
# Random Forest Regression : predict delay in US flights, best results so far
# Boltzmann Machine : estimate delay in arrival flights, accuracy decreases significantly when not in dataset
# Deep learning : Usually involves LSTMs / autoencoders show best results / best approaches use RNNs to reduce # of feature before using SVM or Random Forest to finish the job

# Try : denoising autoencoder (SDA) + Levenberg-Marquart algorithm (LM)

'''
Delay Detection (Classification)
-----------------------------------

1) Random Forest
2) SDA + LM

Delay Estimation (Regression)
-----------------------------------

1) SVR
2) SDA + SVM
'''
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold
from sklearn.svm import SVR,SVC
from sklearn.ensemble import RandomForestRegressor
import pickle
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('../data/raw_data.csv')
df_airports = pd.read_csv('../data/airports.csv')
```

```python
df_for_training = pd.merge(df,df_airports,left_on='airport',right_on='code')
df_for_training.drop(columns=['code','airport_name','carrier','airport'],inplace=True)
df_for_training.rename(columns={'arr_del15':'arr_delay_ct','name':'airport_name','city':'airport_city','state':'airport_state','lat':'airport_lat','lon':'airport_lon'},inplace=True)
df_for_training['carrier_name'] = df_for_training['carrier_name'].apply(lambda x: x.strip())
df_for_training['airport_name'] = df_for_training['airport_name'].apply(lambda x: x.strip())
df_for_training.info()
```

```python
carrier_names = np.unique(df_for_training['carrier_name'].values)
airport_names = np.unique(df_for_training['airport_name'].values)
#city_names = np.unique(df_for_training['airport_city'].values)
#state_names = np.unique(df_for_training['airport_state'].values)

x = pd.DataFrame()
y = pd.DataFrame()

y['carrier_delay_prob'] = df_for_training['carrier_ct']/df_for_training['arr_flights']
y['nas_delay_prob'] = df_for_training['nas_ct']/df_for_training['arr_flights']
#y['human_delay_prob'] = (df_for_training['carrier_ct']+df_for_training['nas_ct'])/df_for_training['arr_flights']
#y['delay_prob'] = df_for_training['arr_delay_ct']/df_for_training['arr_flights']

y['carrier_delay'] = df_for_training['carrier_delay']/df_for_training['carrier_ct']
y['nas_delay'] = df_for_training['nas_delay']/df_for_training['nas_ct']
#y['human_delay'] = (df_for_training['carrier_delay']+df_for_training['nas_delay'])/(df_for_training['carrier_ct']+df_for_training['nas_ct'])
#y['total_delay'] = df_for_training['arr_delay']/df_for_training['arr_delay_ct']

y.replace([np.inf, -np.inf, np.nan], 0, inplace=True)

x['year'] = df_for_training['year']
x['month'] = df_for_training['month']
#x['airport_city'] = df_for_training['airport_city']
#x['airport_state'] = df_for_training['airport_state']
x['airport_name'] = df_for_training['airport_name']
#x['airport_lat'] = df_for_training['airport_lat']
#x['airport_lon'] = df_for_training['airport_lon']
x['carrier_name'] = df_for_training['carrier_name']

x.info()
y.info()

x.to_csv('model_inputs.csv', encoding='utf-8', index=False)
y.to_csv('model_outputs.csv', encoding='utf-8', index=False)
```

```python
def plot_results(model_handle,x,y,y_handle):
    
    enc = OneHotEncoder(handle_unknown='ignore')

    carrier_enc = enc.fit_transform(x[['carrier_name']].values).toarray()
    airport_enc = enc.fit_transform(x[['airport_name']].values).toarray()
    month_enc = enc.fit_transform(x[['month']].values).toarray()
    year_enc = x[['year']].values-2003

    nas_delay_prob = y[['nas_delay_prob']].values
    carrier_delay_prob = y[['carrier_delay_prob']].values
    nas_delay = y[['nas_delay']].values
    carrier_delay = y[['carrier_delay']].values
    
    outputs = {'nas_delay_prob':nas_delay_prob,'nas_delay':nas_delay,
               'carrier_delay_prob':carrier_delay_prob,'carrier_delay':carrier_delay}
    
    train_idxs = (x['year']<2019).values
    test_idxs_2019 = (x['year']==2019).values
    test_idxs_2020 = (x['year']==2020).values
    
    xdata = np.hstack((year_enc,month_enc,airport_enc,carrier_enc))
    ydata = np.squeeze(outputs[y_handle])

    xtrain = xdata[train_idxs]
    ytrain = ydata[train_idxs]

    xtest_2019 = xdata[test_idxs_2019]
    ytest_2019 = ydata[test_idxs_2019]

    xtest_2020 = xdata[test_idxs_2020]
    ytest_2020 = ydata[test_idxs_2020]
    
    models = {}
    for i in range(4):
        with open('../models/rf_model'+str(i)+'.pkl', 'rb') as fp: 
            models[i] = pickle.load(fp)
            
    model = models[model_handle]
    
    fig, ax = plt.subplots(2,2,figsize=(15,10))

    sns.scatterplot(x=ytrain,y=model.predict(xtrain),ax=ax[0,0],alpha=0.6,label='2003-2018 (Train data)')
    ax[0,0].set_ylabel('Predicted delay rate')
    ax[0,0].set_xlabel('Montly delay rate')
    ax[0,0].legend()

    sns.histplot(model.predict(xtrain)-ytrain,kde=True,ax=ax[0,1],label='2003-2018 (Train data)')
    ax[0,1].set_xlim(-0.25,0.25)
    ax[0,1].set_ylabel('Occurrences')
    ax[0,1].set_xlabel('Residual Error')
    ax[0,1].legend()

    sns.scatterplot(x=ytest_2019,y=model.predict(xtest_2019),color='green',alpha=0.6,label='2019 (Test data)',ax=ax[1,0],zorder=1)
    sns.scatterplot(x=ytest_2020,y=model.predict(xtest_2020),color='tomato',alpha=0.6,label='2020 (Test data)',ax=ax[1,0],zorder=2)
    ax[1,0].set_ylabel('Predicted delay rate')
    ax[1,0].set_xlabel('Montly delay rate')
    ax[1,0].legend()

    sns.histplot(model.predict(xtest_2019)-ytest_2019,kde=True,color='green',label='2019 (Test data)',ax=ax[1,1],zorder=2)
    sns.histplot(model.predict(xtest_2020)-ytest_2020,kde=True,color='tomato',label='2020 (Test data)',ax=ax[1,1],zorder=1)
    ax[1,1].set_xlim(-0.25,0.25)
    ax[1,1].set_ylabel('Occurrences')
    ax[1,1].set_xlabel('Residual Error')
    ax[1,1].legend()

    plt.show()
```
